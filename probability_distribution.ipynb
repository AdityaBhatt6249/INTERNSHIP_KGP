```python
import numpy as np
import pandas as pd

# Function to calculate the CDF
def exponential_cdf(x, lambda_param):
    return 1 - np.exp(-lambda_param * np.array(x))

# Assuming you have your dataframe 'df' already loaded

# Function to calculate lambda parameter and extract delta times
def calculate_lambda_param(row_data):
    value_h = row_data.iloc[7]  # Assuming column H is the 8th column
    num_data_points = int(value_h)

    start_column_index = 8  # Assuming delta times start from column I (9th column)
    end_column_index = start_column_index + num_data_points - 1

    delta_times = row_data.iloc[start_column_index:end_column_index + 1].dropna().astype(float).tolist()
    delta_times = [x for x in delta_times if x > 0]

    if not delta_times:
        return None, []

    delta_times.sort()
    lambda_param = 1 / np.mean(delta_times)

    # Calculate lambda values for each inter-arrival time using the CDF values
    lambda_values = []
    cdf_values = exponential_cdf(delta_times, lambda_param)
    for cdf, x in zip(cdf_values, delta_times):
        if 0 < cdf < 1:  # Only consider valid CDF values
            lambda_value = -np.log(1 - cdf) / x
            if np.isfinite(lambda_value):
                lambda_values.append(lambda_value)

    return lambda_param, lambda_values

# Open a text file for writing lambda values
output_file = 'lambda_values_output.txt'
with open(output_file, 'w') as file:
    # Iterate over rows in the dataframe
    for row_number in range(len(df)):
        row_data = df.iloc[row_number]

        # Calculate lambda parameter and lambda values for the current row
        lambda_param, lambda_values = calculate_lambda_param(row_data)

        if lambda_values:
            # Write lambda parameter and lambda values to the file
            file.write(f'Row {row_number + 1}: Lambda Parameter = {lambda_param}, Lambda Values = {lambda_values}\n')

# Inform the user where the output file is saved
print(f'Lambda values saved to {output_file}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate RMS error
def calculate_rms_error(actual_lambda, estimated_lambdas):
    errors = [(actual_lambda - est_lambda) ** 2 for est_lambda in estimated_lambdas]
    mean_squared_error = np.mean(errors)
    return np.sqrt(mean_squared_error)

# Read the lambda values from the output file
output_file = 'lambda_values_output.txt'
lambda_values = []
actual_lambdas = []

with open(output_file, 'r') as file:
    for line in file:
        if line.startswith('Row '):
            # Extract the lambda values from the line
            # Split the line based on 'Lambda Values = ' to isolate the values
            lambda_part = line.split('Lambda Values = ')[1].strip()
            lambda_str = lambda_part.strip('[]').split(',')  # Remove brackets and split
            lambda_values_row = [float(val) for val in lambda_str if val]
            lambda_values.append(lambda_values_row)

            # Extract the actual lambda parameter from the line
            lambda_param_str = line.split('=')[1].split(',')[0].strip() # Split and extract the actual lambda
            actual_lambda = float(lambda_param_str)
            actual_lambdas.append(actual_lambda)

# Calculate RMS errors for each row
rms_errors = [calculate_rms_error(actual, est) for actual, est in zip(actual_lambdas, lambda_values)]

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(rms_errors) + 1), rms_errors, marker='o', linestyle='-', color='b')
plt.xlabel('Row Number')
plt.ylabel('RMS Error')
plt.title('RMS Error for Each Row')
plt.grid(True)
plt.xticks(ticks=range(1,len(rms_errors)+1))
plt.show()
```

```python
import pandas as pd

def check_max_min_ranges(file_path):
    # Load the Excel file
    df = pd.read_excel(file_path, engine='openpyxl')

    # Iterate over each row
    for i, row in df.iterrows():
        row_max = max(row[0], row[1])
        row_min = min(row[0], row[1])

        # Check if this row's max and min fall within the range of any other row
        falls_within_other_row = False
        for j, other_row in df.iterrows():
            if i != j:
                other_max = max(other_row[0], other_row[1])
                other_min = min(other_row[0], other_row[1])
                if other_min <= row_min <= other_max and other_min <= row_max <= other_max:
                    falls_within_other_row = True
                    break

        if not falls_within_other_row:
            return False

    return True

# Example usage
file_path = '/content/output3 (1).xlsx'
result = check_max_min_ranges(file_path)
print(result)
```

```python
from scapy.all import rdpcap

# Function to count UDP packets and calculate average UDP packet size from a specific source IP and port to a specific destination IP and port
def count_udp_packets_and_avg_size(file_path, src_ip, src_port, dst_ip, dst_port):
    packets = rdpcap(file_path)
    packet_count = 0
    total_size = 0

    # Iterate through packets and count sizes for matching UDP packets
    for packet in packets:
        if 'IP' in packet and packet['IP'].src == src_ip and packet['IP'].dst == dst_ip and 'UDP' in packet:
            if packet['UDP'].sport == src_port and packet['UDP'].dport == dst_port:
                packet_count += 1
                total_size += len(packet)

    if packet_count > 0:
        avg_size = total_size / packet_count  # Calculate average UDP packet size
    else:
        avg_size = 0

    return packet_count, avg_size

# Example usage
pcap_file_path = '/mnt/data/skype_chat1a.pcap'  # Replace with the actual path to your pcap file
src_ip = '131.202.240.45'
src_port = 161
dst_ip = '131.202.240.242'
dst_port = 56587

# Count UDP packets and calculate average UDP packet size from specified source IP, port to destination IP, port
packet_count, avg_packet_size = count_udp_packets_and_avg_size(pcap_file_path, src_ip, src_port, dst_ip, dst_port)
print(f"Number of UDP packets from {src_ip}:{src_port} to {dst_ip}:{dst_port}: {packet_count}")
print(f"Average UDP packet size: {avg_packet_size} bytes")
```

```python
import pandas as pd

# Specify the path to your CSV file
csv_file_path = "/content/CICNTTor_browsing.raw.csv"  # Update this path with the actual path to your CSV file
output_csv_file = "/content/drive/MyDrive/browsing_output.csv"  # Path where the output CSV will be saved

# Function to process each line
def process_line(line):
    # Split the line by comma
    fields = line.strip().split(',')
    # Initialize list to collect fields
    selected_fields = []
    # Start processing from the 9th column
    for field in fields[8:]:
        # Stop if we encounter an empty value
        if pd.isna(field) or field.strip() == '':
            break
        selected_fields.append(field)
    return selected_fields

# Read the file line by line
data = []
try:
    with open(csv_file_path, 'r') as file:
        for line_number, line in enumerate(file, 1):
            fields = process_line(line)
            if fields:
                data.append(fields)
            else:
                print(f"Row {line_number}: No relevant fields found")

    # Convert the list of lists to a DataFrame
    df = pd.DataFrame(data)

    # Save the DataFrame to a CSV file
    df.to_csv(output_csv_file, index=False, header=False)
    print(f"\nProcessed data saved to {output_csv_file}")

except FileNotFoundError:
    print(f"File not found: {csv_file_path}")
except Exception as e:
    print(f"An error occurred while reading the file: {e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

# Load the grayscale image (replace with your image path)
image_path = '/content/fc_tor.png'
image_gray = io.imread(image_path, as_gray=True)

# Compute the 2-dimensional FFT of the image
fft_image = np.fft.fft2(image_gray)

# Shift the zero frequency component to the center of the spectrum
fft_image_shifted = np.fft.fftshift(fft_image)

# Compute the magnitude spectrum (absolute value)
magnitude_spectrum = np.abs(fft_image_shifted)

# Calculate the frequency values for the x-axis
rows, cols = image_gray.shape
freq_x = np.fft.fftfreq(cols)
freq_y = np.fft.fftfreq(rows)


# Plot the 1D FFT magnitude vs frequency along columns (x-axis)
plt.figure(figsize=(8, 6))
plt.scatter(freq_x, np.log(magnitude_spectrum[rows // 2, :] + 1), marker='.')
plt.title('Magnitude vs Frequency along columns')
plt.xlabel('Frequency')
plt.ylabel('Magnitude (log scale)')
plt.grid(True)
plt.tight_layout()

# Save the plot as a PNG file
plt.savefig('/content/fc_tor_magnitude_vs_frequency_columns.png', dpi=150)

plt.show()


# Plot the 1D FFT magnitude vs frequency along rows (y-axis)
plt.figure(figsize=(8, 6))
plt.scatter(freq_y, np.log(magnitude_spectrum[:, cols // 2] + 1), marker='.')
plt.title('Magnitude vs Frequency along rows')
plt.xlabel('Frequency')
plt.ylabel('Magnitude (log scale)')
plt.grid(True)
plt.tight_layout()

# Save the plot as a PNG file
plt.savefig('/content/fc_tor_magnitude_vs_frequency_rows.png', dpi=150)

plt.show()
```
